```python
map_searching = folium.Map(location=[47.60621, -122.33207], zoom_start=4)


for (index, row) in searching.iterrows():
    folium.Marker(location=[row.loc['latitude'], row.loc['longitude']], 
                  popup=row.loc['name'] + ' ' + row.loc['category'], 
                  tooltip='click').add_to(map_searching)
    
   
map_searching
```
